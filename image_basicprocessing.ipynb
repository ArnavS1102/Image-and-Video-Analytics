#####  THRESHOLDING

##### ________simple thresholding ________
##### params: image,threshold for classification,max value if more than threshold,type of threshold
##### returns: threshold used,image

```python
t,thresh_binary=cv.threshold(img,127,255,cv.THRESH_BINARY)
t,thresh_binary_inv=cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
t,thresh_trunc=cv.threshold(img,127,255,cv.THRESH_TRUNC)
t,thresh_tozero=cv.threshold(img,127,255,cv.THRESH_TOZERO)
t,thresh_tozero_inv=cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
li=[thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv,thresh_tozero_inv]
for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(li[i],'gray',vmin=0,vmax=255)
```

##### THRESH_BINARY:  if (x,y)> threshold  = max_val else = 0
#####  THRESH_BINARY_INV:  if (x,y)> threshold  = 0 else = max_val
#####  THRESH_TRUNC:  if (x,y)> threshold  = threshold else = (x,y)
#####  THRESH_TOZERO:  if (x,y)> threshold  = threshold else = 0
#####  THRESH_TOZERO_INV:  if (x,y)> threshold  = 0 else = threshold

##### ________adaptive thresholding ________
##### params: image,max_val,adaptive_method,blocksize(size of neighbourhood),c(constant)
##### returns: image

```python
img1=cv.imread("random.jpg",0)
t,sim=cv.threshold(img1,127,255,cv.THRESH_BINARY)
ad_1 = cv.adaptiveThreshold(img1,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,2)
ad_2=cv.adaptiveThreshold(img1,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,11,2)
plt.imshow(img1)
```

##### Non Linear Contrast

```python
input_image = cv.imread("yujiro.jpg")
c=255 / np.log(1 + np.max(input_image))
temp_img= c*(np.log(1+input_image))
```

```python
temp_img = np.array(temp_img, dtype = np.uint8)
```
